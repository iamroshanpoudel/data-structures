# Implementation of Stack using Array

```python
class Stack:
    
    # Constructor
    def __init__(self):
        self.list = []
        self.size = 0
    
    # Push
    def push(self, value):
        self.list.append(value)
        self.size += 1
    
    # Pop
    def pop(self):
        self.size -= 1
        return self.list.pop()
    
    # is Empty?
    def is_empty(self):
        return self.size == 0
    
    # size of the stack
    def size(self):
        return self.size
    
    # String rep of stack
    def __str__(self):
        temp = ""
        for i in range(self.size - 1, -1, -1):
            temp += "|  " + str(self.list[i]) + "  |\n" 
        return temp
        
    
        
```

```python
a = Stack()
for i in range(10):
    a.push(i)
    print(a)
    print(a.is_empty())
for i in range(10):
    a.pop()
    print(a)
```

# Implementation of Stack using Singly Linked List

```python
class Stack:
    
    ''' Singly Linked List class starts here'''
    class SinglyLinkedList:
                
        '''Node class starts here'''
        class Node:
            
            
            # Constructor for Node
            def __init__(self, value, nxt):
                self.value = value
                self.next = nxt                         
        ''' Node class ends here'''
    
        # Constructor for singly linked list
        def __init__(self):
            self.head = None
            self.size = 0
            
        # Add an item to the front of the list
        def push_front(self, val):
            temp = self.Node(val, None)
            if (self.head == None):
                self.head = temp
            else:
                temp.next = self.head
                self.head = temp
            self.size += 1
        
        # Remove an item from the front of the list
        def pop_front(self):
            temp = self.head.value
            self.head = self.head.next
            self.size -= 1
            return temp
    
    '''Singly Linked List class ends here'''
    
    # Constructor for Stack
    def __init__(self):
        self.stack = self.SinglyLinkedList()
        
    # push an item to the top of the stack
    def push(self, val):
        self.stack.push_front(val)
    
    # pop an item from the top of the stack
    def pop(self):
        return self.stack.pop_front()
    
    # is the stack empty?
    def is_empty(self):
        return self.stack.size == 0
    
    # string representation of stack for printing
    def __str__(self):
        temp = ""
        node = self.stack.head
        while (node != None):
            temp = "|  " + str(node.value) + "  |\n"
            node = node.next
        return temp
        
```

```python
a = Stack()
a.push(10)
print(a)
a.push(100)
print(a)
```
